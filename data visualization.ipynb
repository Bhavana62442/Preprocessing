```python
import pandas as pd
import scipy
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\Desktop\data.csv')
print(df.head())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Country', y='Salary', data=df, palette="coolwarm")
plt.title('Salary Distribution by Country')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Country', style='Purchased', s=100)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Purchased', data=df, palette="Set2")
plt.title('Purchase Status Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Country', data=df, palette="Set1")
plt.title('Country Distribution')
plt.show()
```

```python
country_counts = df['Country'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Country Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
corr_matrix = df[['Age', 'Salary']].corr()
print("Correlation Matrix:\n", corr_matrix)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Age and Salary')
plt.show()
```
